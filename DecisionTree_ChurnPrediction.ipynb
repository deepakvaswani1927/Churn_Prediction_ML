```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('BankChurners2.csv')
data
```

```python
data.columns
```

```python
data = data.drop(['CLIENTNUM','Unnamed: 11','Unnamed: 12','Unnamed: 13','Unnamed: 14','Unnamed: 15','Unnamed: 16','Unnamed: 17','Unnamed: 18','Unnamed: 19','Unnamed: 20','Unnamed: 21'],axis = 1)
```

```python
data.info()
```

```python
data['Attrition_Flag'].replace(['Existing Customer','Attrited Customer'],[0,1],inplace=True)
data['Attrition_Flag']
```

```python
data['Gender'].replace(['M','F'],[0,1],inplace=True)
data['Gender']
```

```python
data.info()
```

```python
data['Education_Level'].unique()
```

```python
data.drop(['Card_Category','Income_Category','Education_Level','Marital_Status','Marital_Status'],axis=1,inplace=True)
data.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
x = data.drop(['Attrition_Flag'],axis = 1)
y = data['Attrition_Flag']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state= 42)
```

```python
dt_clf = DecisionTreeClassifier()
dt_clf.fit(x_train,y_train)

tree.plot_tree(dt_clf)
```

```python
y_pred = dt_clf.predict(x_test)

accuracy = dt_clf.score(x_test, y_test)
print("Accuracy:", accuracy)
```
